```python
#Initialisation
grades = [98,76,71,87,83,90,57,79,82,94]
total = 0
grade_count = 0
# processing stage
for grade in grades:
  total += grade
  grade_count += 1
  average = total / grade_count
  print(f'Average up to current grade: {average:.2f}')
```
