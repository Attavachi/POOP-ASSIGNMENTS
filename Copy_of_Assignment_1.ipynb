```python
#Solution for question 2
principal = 172000
boe_rate = 2.25        # Expressed as a percentage
over_boe_rate = 1.49   # Expressed as a percentage

total_annual_rate_percentage = boe_rate + over_boe_rate
annual_rate_decimal = total_annual_rate_percentage / 100

annual_interest = principal * annual_rate_decimal
interest = annual_interest / 12

print(f"The monthly interest payment is: £{interest:.2f}")
```
